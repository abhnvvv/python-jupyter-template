### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt

# Data
x = list(range(11))  # x values from 0 to 10
y = [i**2 for i in x]  # y values as square of x

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Customize the plot
plt.title('Square of Numbers from 0 to 10')
plt.xlabel('X Values')
plt.ylabel('Y = X²')
plt.grid(True)

# Show the plot
plt.show()
```
